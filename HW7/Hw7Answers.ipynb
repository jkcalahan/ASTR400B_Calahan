```python
M31Orbit=np.genfromtxt('Orbit_M31.txt',dtype=None,names=True)
M33Orbit=np.genfromtxt('Orbit_M33.txt',dtype=None,names=True)
M33Analytic=np.genfromtxt('M33Ana',dtype=None,names=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
```

```python
#Seperation of M31 and M33 according to Gurtina's simulation
Sep1=np.sqrt((M33Orbit['x']-M31Orbit['x'])**2+(M33Orbit['y']-M31Orbit['y'])**2+(M33Orbit['z']-M31Orbit['z'])**2)
#Seperation of M31 and M33 according to analytic solution
Sep2=np.sqrt(M33Analytic['x']**2+M33Analytic['y']**2+M33Analytic['z']**2)
```

```python
plt.figure(1,figsize=(8,6))
plt.plot(M31Orbit['t'],Sep1,label='Simulation')
plt.plot(M33Analytic['t'],Sep2,label='Analytic')
plt.title('Seperation of M33 and M31')
plt.legend()
```

```python
#Velocity difference of M31 and M33 according to Gurtina's simulation
Vel1=np.sqrt((M31Orbit['vx']-M33Orbit['vx'])**2+(M31Orbit['vy']-M33Orbit['vy'])**2+(M31Orbit['vz']-M33Orbit['vz'])**2)
#Velocity difference of M31 and M33 according to analytic solution
Vel2=np.sqrt(M33Analytic['vx']**2+M33Analytic['vy']**2+M33Analytic['vz']**2)
```

```python
plt.figure(1,figsize=(8,6))
plt.plot(M31Orbit['t'],Vel1,label='Simulation')
plt.plot(M33Analytic['t'],Vel2,label='Analytic')
plt.title('Velocity Dispersion between M33 and M31')
plt.legend()
```
