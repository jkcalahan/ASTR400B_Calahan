```python
MW=open('MW_lifetime.txt','r')
M31=open('M31_lifetime.txt','r')
M33=open('M33_lifetime.txt','r')
```

```python
import numpy as np
MWdata=np.genfromtxt('MW_lifetime.txt',dtype=None,names=True)
M31data=np.genfromtxt('M31_lifetime.txt',dtype=None,names=True)
M33data=np.genfromtxt('M33_lifetime.txt',dtype=None,names=True)
```

```python
MWx = np.zeros(len(MWdata))
time = np.zeros(len(MWdata))
MWy = np.zeros(len(MWdata))
MWz = np.zeros(len(MWdata))

M31x = np.zeros(len(MWdata))
M31y = np.zeros(len(MWdata))
M31z = np.zeros(len(MWdata))

M33x = np.zeros(len(MWdata))
M33y = np.zeros(len(MWdata))
M33z = np.zeros(len(MWdata))

#Creating arrays of position information for MW and M31
for i in range(len(MWdata)):
    time[i]=MWdata[i][0]
    MWx[i]=MWdata[i][1]
    MWy[i]=MWdata[i][2]
    MWz[i]=MWdata[i][3]
    M31x[i]=M31data[i][1]
    M31y[i]=M31data[i][2]
    M31z[i]=M31data[i][3]
    M33x[i]=M33data[i][1]
    M33y[i]=M33data[i][2]
    M33z[i]=M33data[i][3]

#Find the COM Separation of MW and M31
xdiff1=MWx-M31x
ydiff1=MWy-M31y
zdiff1=MWx-M31z
separation1=np.sqrt(xdiff1**2+ydiff1**2+zdiff1**2)

#Find the COM Separation of M31 and M33
xdiff2=M33x-M31x
ydiff2=M33y-M31y
zdiff2=M33x-M31z
separation2=np.sqrt(xdiff2**2+ydiff2**2+zdiff2**2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure(1,figsize=(8,6))
plt.plot(time,separation1)
plt.ylabel('Separation (kpc)', fontsize=15)
plt.xlabel('Time (Gyr)', fontsize=15)
plt.title("COM Separation of MW and M31", fontsize=20)
#plt.plot([6.3,6.3],[0,250])   <- about when MW and M31 will collide
```

```python
plt.figure(2,figsize=(8,6))
plt.plot(time,separation2)
plt.ylabel('Separation (kpc)', fontsize=15)
plt.xlabel('Time (Gyr)', fontsize=15)
plt.title("COM Separation of M31 and M33", fontsize=20)
```

```python
M33vx = np.zeros(len(MWdata))
M33vy = np.zeros(len(MWdata))
M33vz = np.zeros(len(MWdata))

M31vx = np.zeros(len(MWdata))
M31vy = np.zeros(len(MWdata))
M31vz = np.zeros(len(MWdata))

MWvx = np.zeros(len(MWdata))
MWvy = np.zeros(len(MWdata))
MWvz = np.zeros(len(MWdata))

for i in range(len(MWdata)):
    time[i]=MWdata[i][0]
    MWvx[i]=MWdata[i][4]
    MWvy[i]=MWdata[i][5]
    MWvz[i]=MWdata[i][6]
    M31vx[i]=M31data[i][4]
    M31vy[i]=M31data[i][5]
    M31vz[i]=M31data[i][6]
    M33vx[i]=M33data[i][4]
    M33vy[i]=M33data[i][5]
    M33vz[i]=M33data[i][6]

#Find the Relative Velocity of MW and M31
vxdiff1=MWvx-M31vx
vydiff1=MWvy-M31vy
vzdiff1=MWvx-M31vz
vseparation1=np.sqrt(vxdiff1**2+vydiff1**2+vzdiff1**2)

#Find the Relative Veocity of M31 and M33
vxdiff2=M33vx-M31vx
vydiff2=M33vy-M31vy
vzdiff2=M33vx-M31vz
vseparation2=np.sqrt(vxdiff2**2+vydiff2**2+vzdiff2**2)
```

```python
plt.figure(3,figsize=(8,6))
plt.plot(time,vseparation1)
plt.ylabel('Relative Velocity (km/s)', fontsize=15)
plt.xlabel('Time (Gyr)', fontsize=15)
plt.title("Relative Velocity between MW and M31", fontsize=20)
```

```python
plt.figure(4,figsize=(8,6))
plt.plot(time,vseparation2)
plt.ylabel('Relative Velocity (km/s)', fontsize=15)
plt.xlabel('Time (Gyr)', fontsize=15)
plt.title("Relative Velocity between M31 and M33", fontsize=20)
```
