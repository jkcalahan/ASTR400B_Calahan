```python
#Find the total mass within the limit set
limit = 20 #kpc
steps = 30
snap = 0

#Calculate density profile
density,radii = Density(snap,limit,steps)
```

```python
#Find what kind of power law the density profile follows
a=Slope(.6,20,density,radii)

rtest=np.linspace(0,limit,steps)
dentest=a[1]*rtest**a[0]
```

```python
#Plot Density and Slope results
plt.plot(radii,density)
plt.plot(rtest,dentest,label='power=%.3f'% a[0])
time=(snap*3)/200
plt.title('Time since simulation started: %.2f Gyr ' %time)
plt.legend()
plt.xscale('symlog')
plt.yscale('symlog')
```
