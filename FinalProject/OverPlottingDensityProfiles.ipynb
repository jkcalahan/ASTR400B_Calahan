```python
#Find the total mass within the limit set
limit = 20 #kpc
steps = 30

#Array of snaps that we want to see the density profile at
SNAPS = [0,100,200,300,400,500,600]

#Calculate density profile
density0,radii0 = Density(SNAPS[0],limit,steps)
print('Done with 0')
density1,radii1 = Density(SNAPS[1],limit,steps)
print('Done with 1')
density2,radii2 = Density(SNAPS[2],limit,steps)
print('Done with 2')
density3,radii3 = Density(SNAPS[3],limit,steps)
print('Done with 3')
density4,radii4 = Density(SNAPS[4],limit,steps)
print('Done with 4')
density5,radii5 = Density(SNAPS[5],limit,steps)
print('Done with 5')
density6,radii6 = Density(SNAPS[6],limit,steps)
print('Done Done')
```

```python
plt.plot(radii0,density0,label='Snap 000')
plt.plot(radii1,density1,label='Snap 100')
plt.plot(radii2,density2,label='Snap 200')
plt.plot(radii3,density3,label='Snap 300')
plt.plot(radii4,density4,label='Snap 400')
plt.plot(radii5,density5,label='Snap 500')
plt.plot(radii6,density6,label='Snap 600')
plt.xscale('symlog')
plt.yscale('symlog')
plt.legend()
```
