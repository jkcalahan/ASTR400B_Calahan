```python
#Find the total mass within the limit set
limit = 20 #kpc
steps = 30

#Array of snaps that we want to see the density profile at
SNAPS = [0,100,200,300,400,501,600,700,800]

#Calculate density profile
density0,radii0,t0 = Density(SNAPS[0],limit,steps)
print('Done with 0')
density1,radii1,t1 = Density(SNAPS[1],limit,steps)
print('Done with 1')
density2,radii2,t2 = Density(SNAPS[2],limit,steps)
print('Done with 2')
density3,radii3,t3 = Density(SNAPS[3],limit,steps)
print('Done with 3')
density4,radii4,t4 = Density(SNAPS[4],limit,steps)
print('Done with 4')
density5,radii5,t5 = Density(SNAPS[5],limit,steps)
print('Done with 5')
density6,radii6,t6 = Density(SNAPS[6],limit,steps)
print('Done with 6')
density7,radii7,t7 = Density(SNAPS[7],limit,steps)
print('Done with 7')
density8,radii8,t8 = Density(SNAPS[8],limit,steps)
print('Done with 8')
```

```python
plt.plot(radii0,density0,color='darkblue',label='Snap 000')
plt.plot(radii1,density1,color='mediumblue',label='Snap 100')
plt.plot(radii2,density2,color='blue',label='Snap 200')
plt.plot(radii3,density3,color='royalblue',label='Snap 300')
plt.plot(radii4,density4,color='dodgerblue',label='Snap 400')
plt.plot(radii5,density5,color='deepskyblue',label='Snap 500')
plt.plot(radii6,density6,color='skyblue',label='Snap 600')
plt.plot(radii7,density7,color='lightskyblue',label='Snap 700')
plt.plot(radii7,density7,color='powderblue',label='Snap 800')
plt.xlabel('Time (log(Gyr))')
plt.ylabel('Density Profile (log)')
plt.title('Density Profiles Over Time')
plt.xscale('symlog')
plt.yscale('symlog')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```
