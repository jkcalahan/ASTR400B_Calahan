```python
#Open MassProfiles and create objects for each Galaxy
exec(open('MassProfile.py').read())
MW=MassProfile('MW',0)
M31=MassProfile('M31',0)
M33=MassProfile('M33',0)
print('done running classes')

#Create an array of radii .1 kpc to 30 kpc
radii=np.logspace(-1,1.477121255,20)
```

```python
#Calculate Milky Way mass Components
MWmassdisk = MW.MassEnclosed(2,radii)
print('done with disk')
MWmasshalo = MW.MassEnclosed(1,radii)
print('done with halo')
MWmassbulge = MW.MassEnclosed(3,radii)
print('done with bulge')

#the fuction MassEnclosedTotal does the following, and I don't call it so it doesn't repeat the above steps
MWtotmass=[]
for j in range(len(radii)):
    MWtotmass.append((MWmassdisk[j]+MWmasshalo[j]+MWmassbulge[j]))
    
#Calculate MW velocity Components
MWveldisk = MW.CircularVelocity(2,radii)
print('Done with disk vel')
MWvelhalo = MW.CircularVelocity(1,radii)
print('Done with halo vel')
MWvelbulge = MW.CircularVelocity(3,radii)
print('Done with bulge vel')

#Calculate total circular velocity
MWtotcirc=MW.CircularVelocityTotal(radii)
print('Done with Circular Velocity')
```

```python
#Calculate M31 mass Components
M31massdisk = M31.MassEnclosed(2,radii)
print('done with disk')
M31masshalo = M31.MassEnclosed(1,radii)
print('done with halo')
M31massbulge = M31.MassEnclosed(3,radii)
print('done with bulge')

#the fuction MassEnclosedTotal does the following, and I don't call it so it doesn't repeat the above steps
M31totmass=[]
for j in range(len(radii)):
    M31totmass.append((M31massdisk[j]+M31masshalo[j]+M31massbulge[j]))
    
#Calculate M31 velocity Components
M31veldisk = M31.CircularVelocity(2,radii)
print('Done with disk vel')
M31velhalo = M31.CircularVelocity(1,radii)
print('Done with halo vel')
M31velbulge = M31.CircularVelocity(3,radii)
print('Done with bulge vel')

#Calculate total circular velocity
M31totcirc=M31.CircularVelocityTotal(radii)
print('done with circular velocity')
```

```python
#Calculate M33 Components
M33massdisk = M33.MassEnclosed(2,radii)
print('done with disk')
M33masshalo = M33.MassEnclosed(1,radii)
print('done with halo')
M33massbulge = M33.MassEnclosed(3,radii)
print('done with bulge')

#the fuction MassEnclosedTotal does the following, and I don't call it so it doesn't repeat the above steps
M33totmass=[]
for j in range(len(radii)):
    M33totmass.append((M33massdisk[j]+M33masshalo[j]+M33massbulge[j]))
    
#Calculate MW velocity Components
M33veldisk = M33.CircularVelocity(2,radii)
print('Done with disk vel')
M33velhalo = M33.CircularVelocity(1,radii)
print('Done with halo vel')
M33velbulge = M33.CircularVelocity(3,radii)
print('Done with bulge vel')

#Calculate total circular velocity
M33totcirc=M33.CircularVelocityTotal(radii)
print('Done with Circular Velocity')
```

```python
#Calculate Hernquist things
aMW, aM31, aM33 =60,60,25
MWHernMass = MW.HernquistMass(radii,aMW,max(MWmasshalo))
M31HernMass = M31.HernquistMass(radii,aM31,max(M31masshalo))
M33HernMass = M33.HernquistMass(radii,aM33,max(M33masshalo))

MWHernVel = MW.HernquistVCirc(radii,aMW,max(MWmasshalo))
M31HernVel = M31.HernquistVCirc(radii,aM31,max(M31masshalo))
M33HernVel = M33.HernquistVCirc(radii,aM33,max(M33masshalo))
```

```python
#Import modules that allow me to plot in .ipynb
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#plot the Mass Function of the Milky Way
plt.figure(1,figsize=(8,6))
plt.semilogy(radii,MWmassdisk,'c',label="Disk")
plt.semilogy(radii,MWmasshalo,'m',label="Halo")
plt.semilogy(radii,MWmassbulge,'b',label="Bulge")
plt.semilogy(radii,MWtotmass,'k',label="Total")
plt.semilogy(radii,MWHernMass,'y*',markersize=10,label=('a=',aMW))
plt.ylabel("M(r) (MSun)", fontsize=15)
plt.xlabel("Radius (kpc)", fontsize=15)
plt.title("Mass Fuction of Milky Way Components", fontsize=20)
plt.legend(fontsize=15)
```
